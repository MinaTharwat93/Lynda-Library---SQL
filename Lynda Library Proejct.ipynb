```python
cursor.execute("""select l.BookID,l.PatronID,b.Title,p.FirstName,p.LastName,p.Email,l.LoanDate,l.DueDate from loans as l
inner join patrons as p on l.PatronID=p.PatronID
inner join books as b on l.BookID=b.BookID;
""")
Check_out_Books = cursor.fetchall()
Check_out_Books
```

```python
with open("Check_out_Books.csv", mode='w', encoding='utf-8') as fd:
    fd.write("BookID,PatronID,Title,FirstName,LastName,Email,LoanDate,DueDate\n") 
    for i in range(len(Check_out_Books)):
        BookID = Check_out_Books[i][0]
        PatronID = Check_out_Books[i][1]
        Title = Check_out_Books[i][2]
        FirstName = Check_out_Books[i][3]
        LastName = Check_out_Books[i][4]
        Email = Check_out_Books[i][5]
        LoanDate = Check_out_Books[i][6]
        DueDate = Check_out_Books[i][7]
        fd.write(f"{BookID},{PatronID},\"{Title}\",\"{FirstName}\",\"{LastName}\",\"{Email}\",{LoanDate},{DueDate}\n")

df = pd.read_csv("Check_out_Books.csv")
df
```

```python
cursor.execute("""select l.PatronID,count(l.LoanID) as loan_count ,p.FirstName,p.LastName,p.Email from loans as l
inner join patrons as p on l.PatronID=p.PatronID
group by l.PatronID
order by loan_count asc
limit 10;
""")
report_10_patrons_fewest_books = cursor.fetchall()
report_10_patrons_fewest_books
```

```python
with open("report_10_patrons_fewest_books.csv", mode='w', encoding='utf-8') as fd:
    fd.write("PatronID,LoanCount,FirstName,LastName,Email\n") # header
    for i in range(len(report_10_patrons_fewest_books)):
        PatronID = report_10_patrons_fewest_books[i][0]
        loan_count = report_10_patrons_fewest_books[i][1]
        FirstName = report_10_patrons_fewest_books[i][2]
        LastName = report_10_patrons_fewest_books[i][3]
        Email = report_10_patrons_fewest_books[i][4]
        fd.write(f"{PatronID},{loan_count},\"{FirstName}\",\"{LastName}\",\"{Email}\"\n")

df = pd.read_csv("report_10_patrons_fewest_books.csv")
df
```

```python
cursor.execute("""select b.Title,b.Author,b.Published from books as b 
inner join loans as l on b.BookID=l.BookID
where l.ReturnedDate is not null and b.Published between 1890 and 1899
group by b.Title,b.Author,b.Published
order by b.Published;
""")
available_books_from_1890 = cursor.fetchall()
available_books_from_1890
```

```python
with open("available_books_from_1890.csv", mode='w', encoding='utf-8') as fd:
    fd.write("Title,Author,Published\n") # header
    for i in range(len(available_books_from_1890)):
        Title = available_books_from_1890[i][0]
        Author = available_books_from_1890[i][1]
        Published = available_books_from_1890[i][2]
        fd.write(f"\"{Title}\",\"{Author}\",{Published}\n")

df = pd.read_csv("available_books_from_1890.csv")
df
```

```python
cursor.execute("""SELECT Published, COUNT(DISTINCT(Title)) AS CountPublishedBooks FROM Books
GROUP BY Published
ORDER BY Published ASC;
""")
books_published_each_year = cursor.fetchall()
books_published_each_year
```

```python
with open("books_published_each_year.csv", mode='w', encoding='utf-8') as fd:
    fd.write("Published,CountPublishedBooks\n") # header
    for i in range(len(books_published_each_year)):
        Published = books_published_each_year[i][0]
        CountPublishedBooks = books_published_each_year[i][1]
        fd.write(f"{Published},{CountPublishedBooks}\n")

df = pd.read_csv("books_published_each_year.csv")
df
```

```python
cursor.execute("""SELECT b.Title, b.Author, b.Published, COUNT(l.LoanID) AS count_loans FROM books As b
INNER JOIN loans As l ON b.BookID = l.BookID
GROUP BY b.Title, b.Author, b.Published
ORDER BY count_loans DESC
LIMIT 5;
""")
most_5_popular_books = cursor.fetchall()
most_5_popular_books
```

```python
with open("most_5_popular_books.csv", mode='w', encoding='utf-8') as fd:
    fd.write("Title,Author,Published,count_loans\n") # header
    for i in range(len(most_5_popular_books)):
        Title = most_5_popular_books[i][0]
        Author = most_5_popular_books[i][1]
        Published = most_5_popular_books[i][2]
        count_loans = most_5_popular_books[i][3]
        fd.write(f"{Title},{Author},{Published},{count_loans}\n")

df = pd.read_csv("most_5_popular_books.csv")
df
```
